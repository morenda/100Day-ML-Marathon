# 作業重點

參考步驟: function y=(x+5)² starting from the point x=3

Step 1 : Initialize x =3. Then, find the gradient of the function, dy/dx = 2*(x+5).

Step 2 : Move in the direction of the negative of the gradient. We use a learning rate. Let us assume the learning rate → 0.01

Step 3 : Let’s perform 2 iterations of gradient descent

Step 4 : We can observe that the X value is slowly decreasing and should converge to -5 (the local minima). However, how many iterations should we perform?

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度

## 1. learning rate 0.01 (範例)

```python
cur_x = 3
lr = 0.01 
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0
df = lambda x : 2 * (x + 5)
iters_history = [iters]
x_history = [cur_x]
preci_history = [previous_step_size]
```

```python
while previous_step_size > precision and iters < max_iters :
    prev_x = cur_x
    cur_x = cur_x - lr * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)
    iters = iters + 1
    # print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x, "\nprecision is", previous_step_size) 
    iters_history.append(iters)
    x_history.append(cur_x)
    preci_history.append(previous_step_size)
```

```python
print("Totally Iteraions :", iters)
print("The Local Minumum Occurs at :", cur_x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(iters_history, x_history, 'o-', ms=3, lw=1.5, color = "grey")
plt.xlabel(r'$iters$', fontsize=16)
plt.ylabel(r'$x$', fontsize=16)
plt.title("Graph 1 --- Learning Rate : 0.01")
plt.show()
```

#### 附錄 : 把三種learning rate寫成迴圈

```python
cur_x = 3
LR = [0.00001,0.01,0.1]  
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0
df = lambda x : 2 * (x + 5)
iters_history = [iters]
x_history = [cur_x]
iters_all=[]
x_all=[]
```

```python
for lr in LR:
    while previous_step_size > precision and iters < max_iters :
        prev_x = cur_x
        cur_x = cur_x - lr * df(prev_x)
        previous_step_size = abs(cur_x - prev_x)
        iters = iters + 1
        iters_history.append(iters)
        x_history.append(cur_x)
                  
    iters_all.append(iters_history)
    x_all.append(x_history)
    
    cur_x = 3
    previous_step_size = 1
    iters = 0
    iters_history = [iters]
    x_history = [cur_x]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(figsize=(16, 6))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.plot(iters_all[i], x_all[i], 'o-', ms=3, lw=1, color='blue')
    plt.xlabel(r'$iters$', fontsize = 16)
    plt.ylabel(r'$x$', fontsize = 16)
    plt.title("learning rate= " + str(LR[i]))
plt.show()
```

#### 以第三張圖的learning rate 0.1 的迭代次數最少
